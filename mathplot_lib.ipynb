```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6]
y = [1,21,2,16,7,3]

plt.plot(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('The Chart')
plt.show()
```

```python
x = [1,2,3,4,5,6]
y = [1,21,41,516,212,100]


# Create custom line plot 


plt.plot(x,y, color='green', linestyle='dashed', marker='o')
plt.grid(True)
```

```python
x = [1,2,4,5,6]

y1 = [1,4,9,16,25]
y2 = [1,2,3,4,5]


plt.figure(figsize=(9,5))
# 1,2,1 
# Co 2 cot, 1 hang, vi tri trong hang la thu 1
plt.subplot(1,2,1)
plt.plot(x, y1, color='green')
plt.title('Plot 01')

# 1,2,1 
# Co 2 cot, 1 hang, vi tri trong hang la thu 2
plt.subplot(1,2,2)
plt.plot(x, y2, color='green')
plt.title('Plot 02')
```
